```python
# Loading Libraries 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Loading Cleaned Dataset
df_cleaned = pd.read_csv(r'C:\Users\hp\OneDrive\Desktop\Data_Analyst\Disease_Symptoms_Patients\datasets\Cleaned_Disease_Symptoms_Patients.csv')
df_cleaned
```

```python
# Pie chart for gender distribution of Fever symptom (example)
gender_distribution = df_cleaned.groupby('Fever')['Gender'].value_counts().unstack().fillna(0)
gender_distribution = gender_distribution.loc[1]  # Focus on Fever occurrence (1 for Yes)

# Pie chart visualization
plt.figure(figsize=(6,6))
gender_distribution.plot.pie(autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff'])
plt.title('Gender Distribution of Fever')
plt.ylabel('')
plt.show()
```

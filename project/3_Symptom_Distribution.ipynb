```python
# Loading Libraries 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Loading Cleaned Dataset
df_cleaned = pd.read_csv(r'C:\Users\hp\OneDrive\Desktop\Data_Analyst\Disease_Symptoms_Patients\datasets\Cleaned_Disease_Symptoms_Patients.csv')
df_cleaned
```

```python
# Melt the data to combine symptom columns into one
symptom_columns = ['Fever', 'Cough', 'Fatigue', 'Difficulty Breathing']
symptom_data = df_cleaned.melt(id_vars='Age Group', value_vars=symptom_columns, var_name='Symptom', value_name='Presence')
```

```python
# Plot stacked bar chart
plt.figure(figsize=(10,6))
sns.countplot(x='Age Group', hue='Symptom', data=symptom_data, palette='Set1')
plt.title('Symptom Distribution by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count of Symptoms')
plt.grid(True)
plt.show()
```

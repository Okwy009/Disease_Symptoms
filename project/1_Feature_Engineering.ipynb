```python
# Loading Libraries 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("uom190346a/disease-symptoms-and-patient-profile-dataset")

file_path = r'C:\Users\hp\.cache\kagglehub\datasets\uom190346a\disease-symptoms-and-patient-profile-dataset\versions\2\Disease_symptom_and_patient_profile_dataset.csv'

df = pd.read_csv(file_path)
df.info()
```

```python
# Data Cleanup

# Convert all categorical symptoms to binary values (Yes = 1, No = 0)
symptom_columns = ['Fever', 'Cough', 'Fatigue', 'Difficulty Breathing']
for col in symptom_columns:
    df[col] = df[col].map({'Yes': 1, 'No': 0})

# Check for any outliers in age column
df['Age'].describe()


```

```python
# Example transformation for 'Blood Pressure' and 'Cholesterol Level'
df['Blood Pressure'] = df['Blood Pressure'].map({'Low': 0, 'Normal': 1, 'High': 2})
df['Cholesterol Level'] = df['Cholesterol Level'].map({'Low': 0, 'Normal': 1, 'High': 2})

df.head()
```

- Feature Engineering

```python
# Creating age groups
bins = [0, 18, 35, 50, 65, 100]
labels = ['Child', 'Young Adult', 'Adult', 'Senior', 'Elderly']
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels)

df.head()
```

```python
df.to_csv('Cleaned_Disease_Symptoms_Patients.csv', index=False)
```
